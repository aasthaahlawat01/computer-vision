```python
import numpy as np
import matplotlib.pyplot as plt

import os
import cv2

from PIL import Image

from tensorflow import keras
import tensorflow.keras.layers as k

from tqdm.notebook import tqdm
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
```

```python
classifier=cv2.CascadeClassifier('/Users/aasthaahlawat/Documents/CNN/haarcascade_frontalface_default.xml')

cam=cv2.VideoCapture(0)

while True:
    _,img=cam.read()
    img=cv2.flip(img,1)
    
    try:
     faces=classifier.detectMultiScale(img,2.1,1)
     for(x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,180,0),2)
    except:
       pass

    
    cv2.imshow("Frame",img)
    
    print(faces)
    print('*'*100)

    
    key=cv2.waitKey(30)

    if(key==27):
        cam.release()
        cv2.destroyAllWindows()  
        break
```

SHOWING THE LARGEST FACE AS A SEPERATE FRAME

```python
classifier=cv2.CascadeClassifier('/Users/aasthaahlawat/Documents/CNN/haarcascade_frontalface_default.xml')

cam=cv2.VideoCapture(0)

while True:
    is_image,img=cam.read()

    if not is_image:
        break
       
    img=cv2.flip(img,1)

    faces=classifier.detectMultiScale(img,1.1,5)
    print(faces)

    if(len(faces)>=1):
        largest_face=max(faces,key=lambda f:f[2]*f[3])
        x,y,w,h=largest_face

        cv2.rectangle(img,(x,y),(x+w,y+h),(0,180,0),2)
        face=img[y:y+h,x:x+w]
        cv2.imshow("Face",face)
    
    cv2.imshow("Frame",img)

    key=cv2.waitKey(30)

    if(key==27):
        cam.release()
        cv2.destroyAllWindows()  
        break
```

CIRCLE ON FACE

```python
classifier=cv2.CascadeClassifier('/Users/aasthaahlawat/Documents/CNN/haarcascade_frontalface_default.xml')

cam=cv2.VideoCapture(0)

while True:
    is_image,img=cam.read()

    if not is_image:
        break
       
    img=cv2.flip(img,1)

    faces=classifier.detectMultiScale(img,1.1,5)
    print(faces)

    if(len(faces)>=1):
        largest_face=max(faces,key=lambda f:f[2]*f[3])
        x,y,w,h=largest_face

        circle_x=x+int(w/2)
        circle_y=y+int(h/2)

        cv2.circle(img,(circle_x,circle_y),200,(0,180,0),-1)
        face=img[y:y+h,x:x+w]
        
        img[y:y+h,x:x+w]=face
        cv2.imshow("Face",face)
    
    cv2.imshow("Frame",img)

    key=cv2.waitKey(30)

    if(key==27):
        cam.release()
        cv2.destroyAllWindows()  
        break
```

SAVE ALL FACES ENCOUNTERED

```python
classifier=cv2.CascadeClassifier('/Users/aasthaahlawat/Documents/CNN/haarcascade_frontalface_default.xml')

cam=cv2.VideoCapture(0)

def save(frame,folder):
        if not os.path.exists(folder):
          os.makedirs(folder)
        name_img=len(os.listdir(folder))+1
        name_img=folder+'/IMG'+str(name_img)+'.png'
        cv2.imwrite(name_img,frame)
        print(name_img,'is exported')

while True:
    is_image,img=cam.read()

    if not is_image:
        break
       
    img=cv2.flip(img,1)

    faces=classifier.detectMultiScale(img,1.1,5)
    print(faces)

    key = cv2.waitKey(30) & 0xFF


    for x,y,w,h in faces:

        face=img[y:y+h,x:x+w]
        cv2.imshow("Face",face)
        
        if(key==13):
            save(face,'People')
        elif(key==127):
            pass
        
    
    cv2.imshow("Frame",img)

    if(key==27):
        cam.release()
        cv2.destroyAllWindows()  
        break
```
